Stemming 

```python
from nltk.stem import LancasterStemmer,RegexpStemmer,PorterStemmer,SnowballStemmer
```

```python
l = LancasterStemmer()
r = RegexpStemmer("ing")
p = PorterStemmer()
s = SnowballStemmer("english")
```

```python
l.stem("changing")
```

```python
r.stem("changed")
```

```python
p.stem("change")
```

```python
s.stem("studying")
```
